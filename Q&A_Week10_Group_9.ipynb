**Check the outliers and handle outliers in performance score using Percentiles**

```python
def find_outliers(series):
    Q1 = np.percentile(series, 25, interpolation = 'midpoint')
    Q2 = np.percentile(series, 50, interpolation = 'midpoint')
    Q3 = np.percentile(series, 75, interpolation = 'midpoint')
    IQR = Q3 - Q1
    low_lim = Q1 - 1.5*IQR
    up_lim = Q3 + 1.5*IQR
    print(f'Upper limit = {up_lim} and Lower limit = {low_lim}')
    outlier_list = []
    for x in series:
        if((x>up_lim) or (x<low_lim)):
            outlier_list.append(x)
    print(f'Outliers for {series.name} = {outlier_list}')
    return outlier_list
```

```python
income_outliers = find_outliers(df1['income'])
#df_new.loc[income_outliers].index
income_outliers
```

```python
performance_outliers = find_outliers(df1['performance_score'])
index = df1[df1['performance_score'].isin(performance_outliers)].index
df1.drop(index, inplace = True)
```
